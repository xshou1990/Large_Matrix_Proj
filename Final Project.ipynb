```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy.linalg
from sklearn.model_selection import train_test_split
from timeit import default_timer as timer
```

```python
sparcs = pd.read_csv("~/data/CDPHP/sparcs25%Newborn.csv")
```

```python
tot_charges = pd.read_csv("~/SPARCS_Proj/sparcs25%totcharges.csv") 
```

```python
sparcs['tot_charges']= tot_charges['Total Charges']
```

```python
#sparcs = sparcs.drop('Length_Stay',axis=1)
```

```python
#sparcs = sparcs.drop(sparcs.iloc[:,100:400],axis=1)
```

```python
sparcs1 = sparcs.sample(frac=0.01,random_state=1512)
```

```python
df_train, df_test = train_test_split(sparcs1, test_size=0.2, stratify=sparcs1['Target'],random_state = 1512)
```

```python
Xtr1 = df_train.drop('Target',axis=1)
Xtr = Xtr1.drop('tot_charges',axis=1).values
ytrbi = df_train['Target'].values
ytrcont = df_train['tot_charges'].values
```

```python
from adagrad1 import adagradregL1
from adagrad2 import adagradLS
from adagrad1diag import adagradregL1diag
from adagrad2diag import adagradLSdiag
from subgrad import SGD
from subgrad2 import sgdhingereg
```

```python
lam=0.00001
Tmax =100
delta = 0.00001
eta = 1e-3
batchsize=100
Z=Xtr
y=np.expand_dims(ytrbi,axis=1)
sgdloss,sgdtime = sgdhingereg(Z, y,lam,Tmax,eta*1.0,batchsize)
adaloss,adatime = adagradregL1(Z, y,lam,Tmax,delta,eta*0.8,batchsize)
adalossdiag,adadiagtime = adagradregL1diag(Z, y,lam,Tmax,delta,eta*3.3,batchsize)
```

```python
def reghingeloss1(x):
    n = Z.shape[0]
    d = Z.shape[1]
    bias = np.ones((n,1))
    Znew = np.append(Z,bias,axis=1)
    diff = np.ones((n,1))-  y*(np.matmul(Znew,np.expand_dims(x,axis=1)))
    loss = 1/n*diff[diff>0].sum() + lam*np.linalg.norm(x,ord=1) 
    return loss
```

```python
x1 = np.zeros(Z.shape[1]+1)
BFGS_fun = np.array([])
BFGS_time = np.array([])
for i in range(100):
    start = timer()
    BFGS= scipy.optimize.minimize(reghingeloss1, x1,method='BFGS',options={'maxiter':i})
# print(np.array([BFGS['fun']])
    BFGS_fun = np.append(BFGS_fun,[BFGS['fun']],axis=0)
    time_count = timer() - start
    BFGS_time = np.append(BFGS_time,time_count)
```

```python
L_BFGS_fun = np.array([])
L_BFGS_time = np.array([])
for i in range(100):
    start = timer()
    L_BFGS= scipy.optimize.minimize(reghingeloss1, x1,method='L-BFGS-B',options={'maxiter':i})
# print(np.array([BFGS['fun']])
    L_BFGS_fun = np.append(L_BFGS_fun,[L_BFGS['fun']],axis=0)
    time_count = timer() - start
    L_BFGS_time = np.append(L_BFGS_time,time_count)
```

```python
Tmax =100
delta = 0.0000001
eta = 1e-4
batchsize=100
Z=Xtr
y=np.expand_dims(ytrcont,axis=1)

SGDlossLS,SGDLStime = SGD(Z, y,Tmax,eta*3,batchsize)
adalossLS, adaLStime = adagradLS(Z, y,Tmax,delta,eta*65000000,batchsize)
adalossLSdiag,adaLSdiagtime = adagradLSdiag(Z, y,Tmax,delta,eta*10000000,batchsize)
```

```python
def cost(x):
    n = Z.shape[0]
    d = Z.shape[1]
    bias = np.ones((n,1))
    Znew = np.append(Z,bias,axis=1)
    loss = np.linalg.norm(y-np.matmul(Znew,np.expand_dims(x,axis=1)),ord=2)
    return loss
x1 = np.ones(Z.shape[1]+1)
BFGS_fun_LS1 = np.array([])
BFGS_time_LS1 = np.array([])

for i in range(100):
    start = timer()
    BFGS= scipy.optimize.minimize(cost, x1,method='BFGS',options={'maxiter':i,'eps': 1.4901161193847656e-06})
    #print(np.array([BFGS['fun']])
    BFGS_fun_LS1 = np.append(BFGS_fun_LS1,[BFGS['fun']],axis=0)
    time_count = timer() - start
    BFGS_time_LS1 = np.append(BFGS_time_LS1,time_count)
```

```python
BFGS_fun_LS2 = np.array([])
BFGS_time_LS2 = np.array([])
for i in range(100):
    start = timer()
    BFGS= scipy.optimize.minimize(cost, x1,method='L-BFGS-B',options={'maxiter':i})
# print(np.array([BFGS['fun']])
    BFGS_fun_LS2 = np.append(BFGS_fun_LS2,[BFGS['fun']],axis=0)
    time_count = timer() - start
    BFGS_time_LS2 = np.append(BFGS_time_LS2,time_count)
```

```python
# plot the accuracy vs iteration 
iteration =np.arange(100)
plt.plot(iteration,sgdloss,'-.',iteration,adaloss,'-.' ,iteration,adalossdiag,'-.',iteration,BFGS_fun,'-.',iteration,L_BFGS_fun,'-.')
plt.legend(('SGD','ada full','ada diag','BFGS','L_BFGS'),loc=1)
plt.xlabel("Iteration")
plt.ylabel("Relative Loss")
plt.title("hinge loss with L1 reg, sparse data")
```

```python
# plot the time vs iteration 
iteration =np.arange(100)
plt.plot(iteration,sgdtime,'-.',iteration,adatime,'-.' ,iteration,adadiagtime,'-.',iteration,BFGS_time,'-.',iteration,L_BFGS_time,'-.')
plt.legend(('SGD','ada full','ada diag','BFGS','L_BFGS'),loc=2)
plt.xlabel("Iteration")
plt.ylabel("time")
plt.title("hinge loss with L1 reg, sparse data")
```

```python
# plot the accuracy vs iteration 
iteration =np.arange(100)
plt.plot(iteration,SGDlossLS/SGDlossLS[0],'-.',iteration,adalossLS/adalossLS[0],'-.' ,iteration,adalossLSdiag/adalossLSdiag[0],'-.',iteration,BFGS_fun_LS1/BFGS_fun_LS1[0],'-.',iteration,BFGS_fun_LS2/BFGS_fun_LS1[0],'-.')
plt.legend(('SGD','ada full','ada diag','BFGS','L_BFGS'))
plt.xlabel("Iteration")
plt.ylabel("Relative Loss")
plt.title("least square with no reg on sparse data")
```

```python
# plot time vs iteration 
iteration =np.arange(100)
plt.plot(iteration,SGDLStime,'-.',iteration,adaLStime,'-.' ,iteration,adaLSdiagtime,'-.',iteration,BFGS_time_LS1,'-.',iteration,BFGS_time_LS2,'-.')
plt.legend(('SGD','ada full','ada diag','BFGS','L_BFGS'),loc=2)
plt.xlabel("Iteration")
plt.ylabel("time")
plt.title("least square with no reg, sparse data")
```
